# fill missing value wiht mean and mode

```python
num=df.select_dtypes(include='float64').columns
num
```

```python
for i in num:
  x=df[i].mean()
  df[i].fillna(x,inplace=True)
```

```python
cat_col=df.select_dtypes(include=object).columns
```

```python
for i in cat_col:
  x=df[i].mode()[0]
  df[i].fillna(x,inplace=True)
```
